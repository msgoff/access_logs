```python

import pandas as pd
import re
```

```python
f=open('access.log.1')
data = f.readlines()
f.close()
```

```python
def get_country(ip):
    from geoip import geolite2
    #pip install python-geoip-python3
    #https://pythonhosted.org/python-geoip/
    match = geolite2.lookup(ip)
    if match:
        return match.country
    else:
        return None
```

```python
def filters(df,column,pattern):
    df.loc[:,column] = df.loc[:,column][df.loc[:,column].apply(lambda x:True if not re.findall(pattern,str(x)) else False)]
    df.dropna(subset=[column],inplace=True)
    return df
```

```python
lst = []
issues = []
for line in data:
    resp = re.split('\s+\-\s+\-\s+|\[|\]',line)
    resp = [x.strip().replace('"','') for x in resp if x]
    if len(resp) == 3:
        lst.append(tuple(resp))
    else:
        issues.append(resp)    
```

```python
len(lst)
```

```python
df = pd.DataFrame(lst,columns=['ip','date','route'])
df = filters(df,'route','api/v2')
```

```python
results_df = df.groupby(['ip','date','route']).count().reset_index()
```

```python
results_df['country'] = ''
```

```python
results_df.loc[:,'country']=results_df.apply(lambda x: get_country(x.ip),axis=1)
```
